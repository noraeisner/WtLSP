```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the Fourier series function
def fourier_series(t, *coefficients):
    result = 0.5 * coefficients[0]  # DC component
    num_terms = (len(coefficients) - 1) // 2

    for i in range(1, num_terms + 1):
        result += coefficients[i] * np.cos(2 * np.pi * i * t)
        result += coefficients[i + num_terms] * np.sin(2 * np.pi * i * t)

    return result


# Generate sample time series data
np.random.seed(42)
t = np.linspace(0, 10, 100)
true_coefficients = [1.0, 2.0, 0.5, -1.0, 0.2]  # Example coefficients
true_data = fourier_series(t, *true_coefficients) + 0.2 * np.random.randn(len(t))

# Define the least squares fitting function
def fit_fourier_series(t, y):
    initial_guess = [1.0] * (2 * 2)  # Initial guess for coefficients
    bounds = ([-np.inf] * (2 * 2), [np.inf] * (2 * 2))  # Coefficient bounds
    coefficients, _ = curve_fit(fourier_series, t, y, p0=initial_guess, bounds=bounds)
    return coefficients

# Fit the Fourier series to the data
fit_coefficients = fit_fourier_series(t, true_data)

# Generate the fitted curve
fitted_curve = fourier_series(t, *fit_coefficients)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t, true_data, label='True Data')
plt.plot(t, fitted_curve, label='Fitted Fourier Series', linestyle='--', color='orange')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Least Squares Fitting with Fourier Series')
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit

def fourier_series(x, *params):
    """
    Fourier series function.

    Parameters:
    - x: input data
    - params: Fourier series coefficients (a0, a1, b1, a2, b2, ...)
    """
    a0 = params[0]
    n = (len(params) - 1) // 2
    cos_terms = params[1:1 + n]
    sin_terms = params[1 + n:]

    result = a0 / 2.0

    for i in range(n):
        result += cos_terms[i] * np.cos((i + 1) * x)
        result += sin_terms[i] * np.sin((i + 1) * x)

    return result

def fit_fourier_series(x_data, y_data, n_terms=3):
    """
    Fit a Fourier series to given data using least squares fitting.

    Parameters:
    - x_data: input x data
    - y_data: input y data
    - n_terms: number of Fourier series terms

    Returns:
    - Optimal parameters for the Fourier series
    """
    initial_guess = [0.0] * (2 * n_terms + 1)
    
    # Perform curve fitting using least squares
    params, covariance = curve_fit(fourier_series, x_data, y_data, p0=initial_guess)

    return params

# Example usage:
if __name__ == "__main__":
    # Generate some example data
    x_data = np.linspace(0, 4 * np.pi, 100)
    y_true = 2.0 * np.sin(x_data) + 0.5 * np.cos(2 * x_data) + 1.0 * np.sin(3 * x_data)
    noise = np.random.normal(0, 0.5, len(x_data))
    y_data = y_true + noise
    
    # Fit Fourier series to the data
    optimal_params = fit_fourier_series(x_data, y_data, n_terms=3)
    
    # get the model out
    
    model = fourier_series(x_data, optimal_params)
    
    print (len(x_data), len(optimal_params))
    
    fig, ax = plt.subplots()
    plt.scatter(x_data, y_data)
    #plt.plot(x_data, model)
    plt.show()
    
    print("Optimal parameters:", optimal_params)
```
